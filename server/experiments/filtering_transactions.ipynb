```python
import pandas as pd
```

```python
transcations_df = pd.read_csv("../data/transactions_train.csv")
articles_df = pd.read_csv("../data/filtered_articles5k.csv")
```

```python
acceptable_articles = articles_df['article_id'].unique()
```

```python
filtered_transactions = transcations_df[transcations_df['article_id'].isin(acceptable_articles)]
```

```python
filtered_transactions
```

```python
article_id_to_gender = []

male_articles = []
female_articles = []

for index, row in articles_df.iterrows():
    if row['section_name'] == "Mens Outerwear":
        male_articles.append(row['article_id'])
        # article_id_to_gender.append({row['article_id'] : "Male"})
    else:
        female_articles.append(row['article_id'])
        # article_id_to_gender.append({row['article_id'] : "Female"})
```

```python
male_counter = 0
female_counter = 0

rows_to_drop = []
rows_to_include = []

for index, row in filtered_transactions.iterrows():
    if row['article_id'] in male_articles and male_counter < 5000:
        male_counter += 1
    elif row['article_id'] in female_articles and female_counter < 5000:
        female_counter += 1
    else:
        rows_to_drop.append(index)
```

```python
filtered_transactions10k = filtered_transactions.drop(rows_to_drop)
```
